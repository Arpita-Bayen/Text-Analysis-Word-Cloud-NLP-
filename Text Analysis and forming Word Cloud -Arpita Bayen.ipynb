#Important libraries

```python
import nltk
nltk.download('inaugural')
from nltk.corpus import inaugural
inaugural.fileids();
import string # this is used for string manipulations
from nltk.corpus import stopwords
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import seaborn as sns
```

#Getting Text from file.

```python
df1=inaugural.raw('1941-Roosevelt.txt')
df1
```

```python
df2=inaugural.raw('1961-Kennedy.txt')
df2
```

```python
df3=inaugural.raw('1973-Nixon.txt')
df3
```

#Counting characters

```python
#len()-to calculate the number of characters including special characters.
#str function is to change the integer into string varaiable
print("Total number of characters in Roosevelt's inaugural speech is="+ str(len(df1)))
print("Total number of characters in Kennedy's inaugural speech is="+ str(len(df2)))
print("Total number of characters in Nixon's inaugural speech is="+ str(len(df3)))
```

#Counting Words

```python

nltk.download('punkt')
```

```python
words1= nltk.word_tokenize(df1)
words2= nltk.word_tokenize(df2)
words3= nltk.word_tokenize(df3)
```

```python
print("Total number of words in the Roosevelt's inaugural speech is ="+ str(len(words1)))
print("Total number of words in the Kennedy's inaugural speech is ="+ str(len(words2)))
print("Total number of words in the Nixon's inaugural speech is ="+ str(len(words3)))
```

#Counting sentences

```python
#tokenize the string into sentences.using nltk.sent_tokenize()
print("Total number of sentences in the Roosevelt's inaugural speech is ="+ str(len(nltk.sent_tokenize(df1))))
print("Total number of sentences in the Kennedy's inaugural speech is ="+ str(len(nltk.sent_tokenize(df2))))
print("Total number of sentences in the Nixon's inaugural speech is ="+ str(len(nltk.sent_tokenize(df3))))
```

#Remove all the stopwords from all three speeches.

```python
nltk.download('stopwords')
```

```python
stop_words = nltk.corpus.stopwords.words('english')
```

##Removing stopwords from Roosevelt's speech

```python
filtered_sentence1 = [w for w in words1 if not w.lower() in stop_words]
#with no lower case conversion
filtered_sentence1= []
  
for w in words1:
    if w not in stop_words:
        filtered_sentence1.append(w)
  
print("Before removing stopwords=", words1)
print("After removing stopwords=",  filtered_sentence1)
print("Total number of words in the Roosevelt's inaugural speech after removing stopwords is ="+ str(len(filtered_sentence1)))
```

##Removing stopwords from Kennedy's speech

```python
filtered_sentence2 = [w for w in words2 if not w.lower() in stop_words]
#with no lower case conversion
filtered_sentence2 = []
  
for w in words2:
    if w not in stop_words:
        filtered_sentence2.append(w)
  
print("Before removing stopwords=", words2)
print("After removing stopwords=", filtered_sentence2)
print("Total number of words in the Kennedy's inaugural speech after removing stopwords is ="+ str(len(filtered_sentence2)))
```

##Removing stopwords from Nixon's speech

```python
filtered_sentence3 = [w for w in words3 if not w.lower() in stop_words]
#with no lower case conversion
filtered_sentence3 = []
  
for w in words3:
    if w not in stop_words:
        filtered_sentence3.append(w)
  
print("Before removing stopwords=", words3)
print("After removing stopwords=",filtered_sentence3)
print("Total number of words in the Nixon's inaugural speech after removing stopwords is ="+ str(len(filtered_sentence3)))
```

#Finding most Frequent words

```python
nltk.FreqDist(filtered_sentence1) #it also contains punctuation marks
```

```python
#creating a list of punctuation marks
punct_list=list(string.punctuation)
#adding "--" to the punctuation list
#punct_list=punct_list.extend("--")
```

##Most frequent words used in Roosevelt's inaugural speech

```python
filtered_sentence1= [word for word in filtered_sentence1 if word not in punct_list]
nltk.FreqDist(filtered_sentence1)
```

##Most frequent words used in Kennedy's inaugural speech

```python
filtered_sentence2= [word for word in filtered_sentence2 if word not in punct_list]
nltk.FreqDist(filtered_sentence2)
```

##Most frequent words used in Nixon's inaugural speech

```python
filtered_sentence3= [word for word in filtered_sentence3 if word not in punct_list]
nltk.FreqDist(filtered_sentence3)
```

#Word Cloud

```python
# Word Cloud for Roosevelt's speech

wordcloud1 = WordCloud(stopwords= stop_words, width = 3000, height = 3000, 
                background_color ='black', 
                min_font_size = 10, random_state=100).generate(df1) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud1) 
plt.axis("off")
plt.xlabel('Word Cloud')
plt.tight_layout(pad = 0) 
plt.title("Word Cloud for Roosevelt's inaugural speech ")

print("Word Cloud for Roosevelt's inaugural speech (after cleaning)!!")
plt.show()
```

```python
# Word Cloud for Kennedy's speech

wordcloud2 = WordCloud(stopwords= stop_words, width = 3000, height = 3000, 
                background_color ='black', 
                min_font_size = 10, random_state=100).generate(df2) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud2) 
plt.axis("off")
plt.xlabel('Word Cloud')
plt.tight_layout(pad = 0) 
plt.title("Word Cloud for Kennedy's inaugural speech ")

print("Word Cloud for Kennedy's inaugural speech (after cleaning)!!")
plt.show()
```
